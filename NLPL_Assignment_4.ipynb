```python

# 1) Install dependencies
!pip install -q sentence-transformers scikit-learn pandas pyarrow tqdm matplotlib seaborn

# 2) Imports
import os
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm

# 3) Configuration - update CSV_PATH if needed
CSV_PATH = "/content/saved_dataset.csv"   # change if your file path differs
OUTPUT_DIR = "/content/embeddings_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 4) Load dataset
df = pd.read_csv('/content/saved_dataset.csv')
print("Loaded shape:", df.shape)
print(df.columns.tolist())
# Inspect a bit
display(df.head(3))

# 5) Choose the text column to embed
# You mentioned 'combined_text' exists. If not, change this variable accordingly.
TEXT_COL = "combined_text"
if TEXT_COL not in df.columns:
    raise ValueError(f"Column {TEXT_COL} not found in dataframe. Available cols: {df.columns.tolist()}")

# Fill missing
df[TEXT_COL] = df[TEXT_COL].fillna("").astype(str)

# 6) Load SentenceTransformer model (fast & good)
MODEL_NAME = "all-MiniLM-L6-v2"   # good default. Replace with mpnet, e5, etc. if desired.
print("Loading model:", MODEL_NAME)
model = SentenceTransformer(MODEL_NAME)

# 7) Encode all texts (batch encoding is fast)
texts = df[TEXT_COL].tolist()
print("Number of texts:", len(texts))

# Encode -> returns numpy array shape (n_samples, dim)
embeddings = model.encode(texts, show_progress_bar=True, batch_size=64)
print("Embeddings shape:", embeddings.shape)

# 8) Save embeddings and augmented dataframe
np.save(os.path.join(OUTPUT_DIR, "embeddings.npy"), embeddings)
# Save dataframe without storing big arrays inline; instead keep pointer columns if needed.
df_out = df.copy()
df_out.to_parquet(os.path.join(OUTPUT_DIR, "with_embeddings.parquet"), index=False)
print("Saved embeddings and dataframe to", OUTPUT_DIR)

# 9) Quick downstream example: classification on genre (if available)
# We'll use 'track_genre_clean' if present; else skip.
LABEL_COL = "track_genre_clean"
if LABEL_COL in df_out.columns and df_out[LABEL_COL].nunique() > 1:
    le = LabelEncoder()
    labels = le.fit_transform(df_out[LABEL_COL].fillna("unknown").astype(str))
    X_train, X_test, y_train, y_test = train_test_split(
        embeddings, labels, test_size=0.2, random_state=42, stratify=labels
    )
    clf = LogisticRegression(max_iter=2000)
    print("Training LogisticRegression on embeddings...")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Classification accuracy: {acc:.4f}")
    print("Classification report (labels -> original):")
    print(classification_report(y_test, y_pred, target_names=le.classes_))
else:
    print(f"Label column '{LABEL_COL}' not present or not usable. Skipping classification example.")

# 10) Clustering example (KMeans)
n_clusters = 10
print(f"Clustering into {n_clusters} clusters with KMeans...")
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
cluster_ids = kmeans.fit_predict(embeddings)
df_out["cluster"] = cluster_ids
print("Cluster counts:\n", df_out["cluster"].value_counts().sort_index())

# Save cluster assignments
df_out.to_parquet(os.path.join(OUTPUT_DIR, "with_clusters.parquet"), index=False)

# 11) Similarity / recommendation example function
def get_top_k_similar(idx=None, text_query=None, k=5):
    """
    Provide top-k similar rows to a given index or text_query.
    Either idx (int) OR text_query (str) must be provided.
    """
    if (idx is None) and (text_query is None):
        raise ValueError("Provide idx or text_query.")
    if text_query is not None:
        q_emb = model.encode([text_query])[0]
        sims = cosine_similarity([q_emb], embeddings)[0]
    else:
        sims = cosine_similarity([embeddings[idx]], embeddings)[0]
    topk_idx = np.argsort(sims)[-k-1:-1][::-1]  # exclude itself if searching by idx
    return df_out.iloc[topk_idx].assign(similarity = sims[topk_idx])

# Example use:
print("\nTop-5 similar to row 0:")
display(get_top_k_similar(idx=0, k=5))

# Example by query:
print("\nTop-5 similar to query 'acoustic folk mellow':")
display(get_top_k_similar(text_query="acoustic folk mellow", k=5))

# 12) Visualization: PCA (2D) colored by label or cluster
print("Reducing dimensionality with PCA (2D) for quick visualization...")
pca = PCA(n_components=2, random_state=42)
reduced = pca.fit_transform(embeddings)

df_vis = pd.DataFrame(reduced, columns=["pc1", "pc2"])
if LABEL_COL in df_out.columns:
    df_vis["label"] = df_out[LABEL_COL].astype(str).fillna("unknown")
else:
    df_vis["label"] = df_out["cluster"].astype(str)

plt.figure(figsize=(10,6))
# Plot sample if too many points
SAMPLE_N = 2000
if len(df_vis) > SAMPLE_N:
    sample = df_vis.sample(SAMPLE_N, random_state=42)
else:
    sample = df_vis
sns.scatterplot(data=sample, x="pc1", y="pc2", hue="label", legend=False, s=10)
plt.title("PCA 2D of embeddings (color = label/cluster)")
plt.show()

# 13) Optional: t-SNE (slower) - enable only if dataset small or you want it
DO_TSNE = False
if DO_TSNE:
    print("Running t-SNE (this may take time)...")
    tsne = TSNE(n_components=2, perplexity=40, n_iter=800, random_state=42)
    reduced_tsne = tsne.fit_transform(embeddings)
    df_tsne = pd.DataFrame(reduced_tsne, columns=["x", "y"])
    df_tsne["label"] = df_out[LABEL_COL].astype(str) if LABEL_COL in df_out.columns else df_out["cluster"].astype(str)
    plt.figure(figsize=(10,6))
    sample = df_tsne.sample(min(2000, len(df_tsne)), random_state=42)
    sns.scatterplot(data=sample, x="x", y="y", hue="label", legend=False, s=10)
    plt.title("t-SNE visualization")
    plt.show()

print("All done. Outputs are under:", OUTPUT_DIR)
```
